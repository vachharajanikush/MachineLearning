```python
# Line Plot
sns.lineplot(df['CreditScore'],df['Age'])
```

```python
# Histogram
sns.distplot(df["CreditScore"].head(20))
```

```python
sns.distplot(df['CreditScore'].head(20),kde=False,bins=10)

# bins etle aa je bars hoy e
#  a histogram with a kernel density estimate (KDE) plot by default.
#However, the 'kde=False' parameter in this code turns off the KDE plot and only displays the histogram.
#The 'bins=10' parameter specifies that the histogram should have 10 bins.
```

```python
# Box Plot 
sns.boxplot(df['CreditScore'].head(20))
```

```python
# Violin Plot
sns.violinplot(df['CreditScore'].head(20))
```

```python
# Swarn Plot
sns.swarmplot(df['CreditScore'].head(20),df["Age"].head(20))
```

```python
# Point Plot
sns.pointplot(df['CreditScore'].head(20),df['Age'].head(20))
```

```python
# Count Plot - used to show count of observations in each categorical bin using bars
sns.countplot(df['CreditScore'])
```

```python
#KDE Plot
# The kernel density estimate is a non-parametric way to estimate the probability density function of a random variable.
#It is based on the idea of representing a probability distribution as a sum of smooth and simple kernel functions,
# centered at each observation.

sns.kdeplot(df['CreditScore'].head(20),df['Age'].head(20))
```
